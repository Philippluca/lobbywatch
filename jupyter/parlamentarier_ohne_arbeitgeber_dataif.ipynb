```python
import datetime, re, requests
```

[Dokumentation der Lobbywatch Datenschnittstelle (dataIF)](https://github.com/lobbywatch/lobbywatch/blob/master/drupal/lobbywatch/lobbywatch_data/data_interface_description.adoc)

```python
d = requests.get("http://cms.lobbywatch.ch/de/data/interface/v1/json/table/parlamentarier/flat/list?limit=none", headers={"Accept":"application/json"}).json()
```

```python
d
```

Generiere **CSV mit allen Berufen/Arbeitgebern**
Flachgekloppt für jeden Beruf/Arbeitgeber gibt es eine neue Zeile. Parlamentarier können so mehrfach vorkommen.

```python
print("Parlamentarier; id; Rat; Kanton; Partei; Fraktion; Beruf; Arbeitgeber; Jobtitel; Start; End")
for o in d['data']:
    parlamentarier = "{}; {}; {}; {}; {}; {}".format(o['anzeige_name'], o['id'], o['rat'], o['kanton'], o['partei'], o['fraktion'])
    if o['parlament_beruf_json'] is not None:
        for b in o['parlament_beruf_json']:
            print("{}; {}; {}; {}; {}; {}".format(parlamentarier, b['beruf'], b['arbeitgeber'], b['jobtitel'], b['von'], b['bis']))
```

Generiere **CSV mit aller Parlamentarier ohne Arbeitgeber**

```python
print("Parlamentarier; id; Rat; Kanton; Partei; Fraktion; Berufe")
for o in d['data']:
    parlamentarier = "{}; {}; {}; {}; {}; {}".format(o['anzeige_name'], o['id'], o['rat'], o['kanton'], o['partei'], o['fraktion'])
    hat_arbeitgeber = False
    berufe = []
    if o['parlament_beruf_json'] is not None:
        for b in o['parlament_beruf_json']:
            hat_arbeitgeber = b['arbeitgeber'] is not None
            if b['beruf'] is not None:
                berufe.append(b['beruf'])
    if not hat_arbeitgeber:
        print("{}; {}".format(parlamentarier, ' / '.join(berufe)))
```
